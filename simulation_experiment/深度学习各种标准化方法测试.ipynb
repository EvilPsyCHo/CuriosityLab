```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x = torch.rand(4, 2, 16)

bn= torch.nn.BatchNorm1d(2)
x_bn = bn(x)

ln = torch.nn.LayerNorm([2, 16])
x_ln = ln(x)

inn = torch.nn.InstanceNorm1d(16)
x_inn = inn(x)
```
